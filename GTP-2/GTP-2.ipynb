Problema 2: Representación de la fecha de nacimiento en la clase Persona

```python
import datetime
class Persona:
    def __init__(self, nombre, apellido, edad, fecha_nacimiento):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad
        self.fec_nac = fecha_nacimiento

    def _calcularedad(self):
        hoy = datetime.date.today()
        edad = hoy.year - self.fec_nac.year
        if (hoy.month, hoy.day) < (self.fec_nac.month, self.fec_nac.day):
            edad -= 1
        return edad
    
    def mostrar(self):
        edad = self._calcularedad()
        hoy = datetime.date.today()
        cumple_actual = datetime.date(hoy.year, self.fec_nac.month, self.fec_nac.day)
        print(f"{self.apellido}, {self.nombre} : {edad} años. Dia del cumpleaños: {cumple_actual}")

class Principal:
    def __init__(self):
        pass
    def ejecutar(self):
        fec_nac = datetime.date(2006, 6, 7)
        persona1 = Persona("Rafael", "Gorrochategui", 19, fec_nac)
        print(f"ID de persona1: {id(persona1)}")
        persona1.mostrar()

if __name__ == "__main__":
    principal = Principal()
    principal.ejecutar()
```

Problema 4: Hashing

```python
import hashlib
class Persona:
    def __init__(self, nombre, apellido, fecha_nacimiento, password):
        self.nombre = nombre
        self.apellido = apellido
        self.fec_nac = fecha_nacimiento
        self.password = hashlib.sha256(password.encode()).hexdigest()

    def _calcularedad(self):
        hoy = datetime.date.today()
        edad = hoy.year - self.fec_nac.year
        if (hoy.month, hoy.day) < (self.fec_nac.month, self.fec_nac.day):
            edad -= 1
        return edad
    
    def mostrar(self):
        edad = self._calcularedad()
        hoy = datetime.date.today()
        cumple_actual = datetime.date(hoy.year, self.fec_nac.month, self.fec_nac.day)
        print(f"{self.apellido}, {self.nombre} : {edad} años. Dia del cumpleaños: {cumple_actual}")

    def validar_password(self, password_v):
        contra1 = hashlib.sha256(password_v.encode()).hexdigest()
        if contra1==self.password:
            return True
        else:
            return False

if __name__ == "__main__":
    juan = Persona("Juan", "Perez", "1990-11-11", "password")
    print(juan.validar_password("password"))  # Salida: True
    print(juan.validar_password("incorrecto"))  # Salida: False
```
