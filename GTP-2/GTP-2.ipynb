Problema 2: Representación de la fecha de nacimiento en la clase Persona

```python
import datetime
class Persona:
    def __init__(self, nombre, apellido, edad, fecha_nacimiento):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad
        self.fec_nac = fecha_nacimiento

    def _calcularedad(self):
        hoy = datetime.date.today()
        edad = hoy.year - self.fec_nac.year
        if (hoy.month, hoy.day) < (self.fec_nac.month, self.fec_nac.day):
            edad -= 1
        return edad
    
    def mostrar(self):
        edad = self._calcularedad()
        hoy = datetime.date.today()
        cumple_actual = datetime.date(hoy.year, self.fec_nac.month, self.fec_nac.day)
        print(f"{self.apellido}, {self.nombre} : {edad} años. Dia del cumpleaños: {cumple_actual}")

class Principal:
    def __init__(self):
        pass
    def ejecutar(self):
        fec_nac = datetime.date(2006, 6, 7)
        persona1 = Persona("Rafael", "Gorrochategui", 19, fec_nac)
        print(f"ID de persona1: {id(persona1)}")
        persona1.mostrar()

if __name__ == "__main__":
    principal = Principal()
    principal.ejecutar()
```

Problema 4: Hashing

```python
import hashlib
import datetime
class Persona:
    def __init__(self, nombre, apellido, fecha_nacimiento, password):
        self.nombre = nombre
        self.apellido = apellido
        self.fec_nac = fecha_nacimiento
        self.password = hashlib.sha256(password.encode()).hexdigest()

    def _calcularedad(self):
        hoy = datetime.date.today()
        edad = hoy.year - self.fec_nac.year
        if (hoy.month, hoy.day) < (self.fec_nac.month, self.fec_nac.day):
            edad -= 1
        return edad
    
    def mostrar(self):
        edad = self._calcularedad()
        hoy = datetime.date.today()
        cumple_actual = datetime.date(hoy.year, self.fec_nac.month, self.fec_nac.day)
        print(f"{self.apellido}, {self.nombre} : {edad} años. Dia del cumpleaños: {cumple_actual}")

    def validar_password(self, password_v):
        contra1 = hashlib.sha256(password_v.encode()).hexdigest()
        if contra1==self.password:
            return True
        else:
            return False

if __name__ == "__main__":
    juan = Persona("Juan", "Perez", "1990-11-11", "password")
    print(juan.validar_password("password"))  # Salida: True
    print(juan.validar_password("incorrecto"))  # Salida: False
```

Problema 6: Sistema de nómina de trabajadores

```python
class Trabajador:
    def __init__(self, nombre, dni, direccion, jefe=None):
        self.nombre = nombre
        self.dni = dni
        self.direccion = direccion
        self.jefe = jefe
    def calcular_remuneracion(self):
        raise NotImplementedError()

class Mensualizado(Trabajador):
    def __init__(self, nombre, dni, direccion, categoria_monto, jefe):
        super().__init__(nombre, dni, direccion, jefe)
        self.monto_fijo = categoria_monto
    def calcular_remuneracion(self):
        return self.monto_fijo

class Jornalizado(Trabajador):
    def __init__(self, nombre, dni, direccion, precio_hora, precio_extra, jefe):
        super().__init__(nombre, dni, direccion, jefe)
        self.precio_hora = precio_hora
        self.precio_extra = precio_extra
        self.horas_trabajadas = 0
    def set_horas(self, horas):
        self.horas_trabajadas = horas
    def calcular_remuneracion(self):
        if self.horas_trabajadas <= 40:
            return self.horas_trabajadas * self.precio_hora
        else:
            return (40 * self.precio_hora) + ((self.horas_trabajadas - 40) * self.precio_extra)

class Jefe(Trabajador):
    def __init__(self, nombre, dni, direccion, salario_acordado):
        super().__init__(nombre, dni, direccion, None)
        self.salario_acordado = salario_acordado
    def calcular_remuneracion(self):
        return self.salario_acordado


if __name__ == "__main__":
    jefe_it = Jefe("Rafael", "12345678", "Santa Fe", 500000)
    emp1 = Mensualizado("Ana", "40123456", "Calle 1", 150000, jefe_it)
    emp2 = Jornalizado("Pedro", "41123456", "Calle 2", 1000, 1500, jefe_it)

    emp2.set_horas(45)

    nomina = [jefe_it, emp1, emp2]

    for t in nomina:
        print(f"{t.nombre}: ${t.calcular_remuneracion()}")
```

Problema 9: Egresados

```python
from typing import List, Optional
from datetime import datetime

class Examen:
    def __init__(self, nota: int, fecha: datetime):
        self.nota = nota
        self.fecha = fecha
    def es_aprobado(self, nota_corte: int = 6) -> bool:
        return self.nota >= nota_corte

class Alumno:
    def __init__(self, nombre: str, fecha_ingreso: datetime, fecha_egreso: Optional[datetime] = None):
        self.nombre = nombre
        self.fecha_ingreso = fecha_ingreso
        self.fecha_egreso = fecha_egreso
        self.examenes: List[Examen] = []
    def es_egresado(self) -> bool:
        return self.fecha_egreso is not None
    def obtener_promedio_aprobados(self) -> float:
        notas = [e.nota for e in self.examenes if e.es_aprobado()]
        return sum(notas) / len(notas) if notas else 0.0
    def obtener_datos_egresado(self) -> str:
        return f"Egresado {self.nombre} - Promedio: {self.obtener_promedio_aprobados()}"

class Carrera:
    def __init__(self, nombre: str):
        self.nombre = nombre
        self.alumnos: List[Alumno] = []
    def listar_egresados(self) -> int:
        total_egresados = 0
        reporte = []
        for alumno in self.alumnos:
            if alumno.es_egresado():
                total_egresados += 1
                reporte.append(alumno.obtener_datos_egresado())
        if reporte:
            print("\n".join(reporte))
        return total_egresados
```

Problema 10: Cátedras

```python
from datetime import datetime
from typing import List

class Persona:
    def __init__(self, nombre: str, dni: str):
        self.nombre = nombre
        self.dni = dni

class Docente(Persona):
    def __init__(self, nombre: str, dni: str):
        super().__init__(nombre, dni)

class Alumno(Persona):
    def __init__(self, nombre: str, dni: str):
        super().__init__(nombre, dni)
        self.carreras: List['Carrera'] = []

class Comision:
    def __init__(self, letra: str, fecha_inicio: datetime, titular: Docente):
        self.letra = letra
        self.fecha_inicio = fecha_inicio
        self.titular = titular
        self.alumnos: List[Alumno] = []
    def es_del_anio(self, anio: int) -> bool:
        return self.fecha_inicio.year == anio
    def obtener_nombres_alumnos(self) -> List[str]:
        # Tell, Don't Ask: La comisión conoce a sus alumnos
        return [f"{a.nombre} (DNI: {a.dni})" for a in self.alumnos]

class Catedra:
    def __init__(self, nombre: str):
        self.nombre = nombre
        self.comisiones: List[Comision] = []
    def obtener_alumnos_por_anio(self, anio: int) -> List[str]:
        # Ley de Demeter: Cátedra solo habla con sus Comisiones
        nombres = []
        for comision in self.comisiones:
            if comision.es_del_anio(anio):
                nombres.extend(comision.obtener_nombres_alumnos())
        return nombres

class Universidad:
    def __init__(self, nombre: str):
        self.nombre = nombre
        self.catedras: List[Catedra] = []
    def listar_alumnos(self, catedra: Catedra):
        anio_actual = datetime.now().year
        print(f"Alumnos en la Cátedra: {catedra.nombre} ({anio_actual})")
        print("-" * 40)
        # Punto único de salida: acumulamos y mostramos
        lista_nombres = catedra.obtener_alumnos_por_anio(anio_actual)
        if lista_nombres:
            for nombre in sorted(set(lista_nombres)): # set para no repetir si cursan 2 veces
                print(nombre)
        else:
            print("No hay alumnos registrados en comisiones de este año.")
```
