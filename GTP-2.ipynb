Problema 1: Identidad de un objeto

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        
class Principal:
    def __init__(self):
        pass
    def ejecutar(self):
        persona1 = Persona("Rafael", 19)
        persona2 = persona1
        persona3 = Persona("Alguien", 19)
        print(f"Identidad de persona1: {id(persona1)}")
        print(f"Identidad de persona2: {id(persona2)}")
        print(f"Identidad de persona3: {id(persona3)}")

        print("\nVerificaciones de Identidad")
        print(f"¿persona1 es la misma que persona2?: {persona1 is persona2}")
        print(f"¿persona1 es la misma que persona3?: {persona1 is persona3}")
  
if __name__ == "__main__":
    principal = Principal()
    principal.ejecutar()
```

Problema 2: Representación de la fecha de nacimiento en la clase Persona

```python
import datetime
class Persona:
    def __init__(self, nombre, apellido, edad, fecha_nacimiento):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad
        self.fec_nac = fecha_nacimiento

    def _calcularedad(self):
        hoy = datetime.date.today()
        edad = hoy.year - self.fec_nac.year
        if (hoy.month, hoy.day) < (self.fec_nac.month, self.fec_nac.day):
            edad -= 1
        return edad
    
    def mostrar(self):
        edad = self._calcularedad()
        hoy = datetime.date.today()
        cumple_actual = datetime.date(hoy.year, self.fec_nac.month, self.fec_nac.day)
        print(f"{self.apellido}, {self.nombre} : {edad} años. Dia del cumpleaños: {cumple_actual}")

class Principal:
    def __init__(self):
        pass
    def ejecutar(self):
        fec_nac = datetime.date(2006, 6, 7)
        persona1 = Persona("Rafael", "Gorrochategui", 19, fec_nac)
        print(f"ID de persona1: {id(persona1)}")
        persona1.mostrar()

if __name__ == "__main__":
    principal = Principal()
    principal.ejecutar()
```
