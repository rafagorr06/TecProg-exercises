## PARTE 1

1- Calculadora de edad

```python
anio_nacimiento = int(input("Introduce tu año de nacimiento: "))
anio_actual = int(input("Introduce el año actual: "))

edad = anio_actual - anio_nacimiento

print("Tu edad es de {edad} años.")
```

2- Conversión de Segundos

```python
segundos_totales = int(input("Introduce la cantidad de segundos: "))
horas = segundos_totales // 3600
segundos_restantes = segundos_totales % 3600
minutos = segundos_restantes // 60
segundos_finales = segundos_restantes % 60
print(f"{segundos_totales} segundos equivalen a: {horas}h {minutos}m {segundos_finales}s")
```

3- Calculadora de Raíces Cuadradas

```python
import math
numero = float(input("Introduce un número para calcular su raíz (negativo para salir): "))
while numero >= 0:
    raiz = math.sqrt(numero)
    print(f"La raíz cuadrada de {numero} es: {raiz:.2f}")
    numero = float(input("Introduce otro número: "))
print("Programa finalizado por ingreso de número negativo.")
```

4- Clasificación de Consumo Eléctrico

```python
kwh = float(input("Ingrese kWh consumidos: "))
if kwh <= 100:
    costo = kwh * 200
    categoria = "Bajo"
elif kwh <= 300:
    costo = kwh * 300
    categoria = "Medio"
else:
    costo = kwh * 400
    categoria = "Alto"
print(f"Categoría: {categoria}")
print(f"Total a pagar: ${costo}")
```

5- Conversión de Unidades

```python
celsius = float(input("Ingrese temperatura en Celsius: "))
fahrenheit = (celsius * 1.8) + 32
print(f"{celsius}°C equivalen a {fahrenheit}°F")
```

6- Cálculo de Distancia de Viaje

```python
vel_prom = int(input("Introduce la velocidad prom.: "))
tiempo_viaje = int(input("Introduce el tiempo de viaje total: "))
dist_tot = vel_prom * tiempo_viaje
print(f"La dist. total es: {dist_tot}")
```

7- Presupuesto de Viaje

```python
costo_alq = int(input("Introduce el costo de alq. del vehículo: "))
num_dias = int(input("Introduce el número de días del alquiler: "))
presup_nafta = int(input("Introduce el presupuesto para la nafta: "))
costo_tot_auto = costo_alq * num_dias
costo_tot = costo_tot_auto + presup_nafta
print(f"El costo total del viaje es de: {costo_tot}")
```

8- Volumen de un Cilindro 

```python
import math
r_base = float(input("Introduce el radio de la base: "))
altura = float(input("Introduce la altura: "))
vol_cilindro = math.pi * (r_base ** 2) * altura
print(f"El volumen total es de: {vol_cilindro}")
```

2- Suma de Números

```python
n = int(input("¿Cuántos números vas a ingresar?: "))
suma_total = 0
for i in range(n):
    numero = float(input(f"Introduce el número {i + 1}: "))
    suma_total = suma_total + numero
print(f"La suma total es: {suma_total}")
```

3- Media de Números

```python
n = int(input("¿Cuántos números vas a ingresar?: "))
suma_total = 0
for i in range(n):
    numero = float(input(f"Introduce el número {i + 1}: "))
    suma_total = suma_total + numero
media = suma_total / n
print(f"La media es: {media}")
```

4- Encontrar el Máximo

```python
n = int(input("¿Cuántos números vas a ingresar?: "))
maximo = float('-inf') 
for i in range(n):
    numero = float(input(f"Introduce el número {i + 1}: "))
    if numero > maximo:
        maximo = numero
print(f"El número máximo es: {maximo}")
```

5- Tabla de Multiplicar

```python
n = int(input("¿De qué número quieres ver la tabla?: "))
print(f"Tabla de multiplicar del {n}:")
for i in range(1, 11):
    resultado = n * i
    print(f"{n} x {i} = {resultado}")
```

6- Suma de Números Pares hasta N

```python
n = int(input("Introduce el número N: "))
suma_pares = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        suma_pares += i
print(f"La suma de los pares hasta {n} es: {suma_pares}")
```

7- Contador de Dígitos

```python
n = int(input("Introduce un número entero: "))
cantidad_digitos = len(str(abs(n)))
print(f"El número tiene {cantidad_digitos} dígitos.")
```

8- Conversión Binaria

```python
n = int(input("Introduce un número para pasar a binario: "))
if n == 0:
    print("Binario: 0")
else:
    binario = ""
    temp_n = n
    while temp_n > 0:
        residuo = temp_n % 2
        binario = str(residuo) + binario
        temp_n = temp_n // 2
    print(f"El número {n} en binario es: {binario}")
```
